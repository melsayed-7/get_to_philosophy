```

import time
import urllib
import bs4
from bs4 import BeautifulSoup
import requests


start_url = "https://en.wikipedia.org/wiki/special:Random"
target_url = "https://en.wikipedia.org/wiki/Philosophy"

def find_first_link(url):
    r = requests.get(url)
    html = r.text
    s = bs4.BeautifulSoup(html, "html.parser")

    content_div = s.find(class_= "mw-parser-output")


    link = None

    for element in content_div.find_all("p", recursive=False):
        print(element)
        if element.find("a", recursive=False):
            link = element.find("a", recursive=False).get('href')
            break

    if not link:
        return

    first_link = urllib.parse.urljoin('https://en.wikipedia.org/', link)
    return first_link

def continue_or_not(jumps, target_url, max_steps=50):
    if jumps[-1] == target_url:
        print("Target article found!")
        return False
    elif len(jumps) > max_steps:
        print("max jumbs reached. can not search more")
        return False
    elif jumps[-1] in jumps[:-1]:
        print("We've arrived at an article we've already seen, aborting search!")
        return False
    else:
        return True

article_chain = [start_url]

while continue_or_not(article_chain, target_url):
    print(article_chain[-1])

    first_link = find_first_link(article_chain[-1])
    if not first_link:
        print("We've arrived at an article with no links, aborting search!")
        break

    article_chain.append(first_link)
    time.sleep(0.5) 
```
